```python
def collect_ratios(tickers: list, ratios: list):
    rows = []
    for ticker in tickers:
        info = yf.Ticker(ticker).info
        row = [ticker] + [info.get(ratio, None) for ratio in ratios]
        rows.append(row)
    return pd.DataFrame(rows, columns=['Ticker'] + ratios)

collect_ratios(['WMT', 'MO', 'NVDA', 'CRM'], ['sector', 'industry', 'currentRatio', 'quickRatio'])
```

```python
# the explicit end date could not be set because of when the market opens/close
hist_prices_2023 = yf.Tickers(['AAPL', 'SMR', 'KO']).history(start= '2023-01-01', period='12mo', interval='1d')
hist_prices_2023
```

```python
def plot_closing_prices(data):
    close_prices = data['Close']
    price_change_in_percentage = (close_prices / close_prices.iloc[0] * 100)
    price_change_in_percentage.plot(figsize = (12, 8), fontsize = 12)
    plt.ylabel('Percentage')
    plt.title(f'Price Chart', fontsize = 15)
    plt.show()

plot_closing_prices(hist_prices_2023)
```

```python
# the explicit end date could not be set because of when the market opens/close
hd_nvda_smr_aapl_1yr = yf.Tickers(['AAPL', 'SMR', 'KO']).history(start='2023-11-22', period='12mo', interval='1d')
plot_closing_prices(hd_nvda_smr_aapl_1yr)
```

```python
hist_prices_2024 = hd_nvda_smr_aapl_1yr
simple_returns = hist_prices_2024['Close'].pct_change(fill_method=None).dropna()
simple_returns
```
