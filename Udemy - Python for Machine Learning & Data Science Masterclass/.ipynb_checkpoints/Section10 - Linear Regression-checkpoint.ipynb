```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df['total_spend'] = df['TV'] + df['radio'] + df['newspaper']
```

```python
# sns.scatterplot(data=df, x='total_spend', y='sales')
sns.regplot(data=df, x='total_spend', y='sales')
```

```python
X = df['total_spend']
y = df['sales']

# y= B1x + B0
b1, b0 = np.polyfit(X, y, deg=1)
np.polyfit(X, y, deg=1)
```

```python
potential_spend = np.linspace(0, 500, 100)
predicted_sales = b1 * potential_spend + b0

sns.regplot(data=df, x='total_spend', y='sales')
plt.plot(potential_spend, predicted_sales, color='red')
```

```python
spend = 200
predicted_sales = b1 * spend + b0
predicted_sales
```

```python
B3, B2, B1, B0 = np.polyfit(X, y, 3)
pot_spend = np.linspace(0, 500, 100)
pred_sales = B3 * pot_spend ** 3 + B2 * pot_spend ** 2 + B1 * pot_spend + B0

sns.regplot(data=df, x='total_spend', y='sales')
plt.plot(pot_spend, pred_sales, color='red')
```

### Scikit-learn

```python
sns.pairplot(df)
```

```python
# generate your features
X = df.drop('sales', axis=1)
y = df['sales']


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

# create the estimator
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(X_train, y_train)
test_predictions = model.predict(X_test)

from sklearn.metrics import mean_absolute_error, mean_squared_error, root_mean_squared_error
mean_absolute_error(y_test, test_predictions)
mean_squared_error(y_test, test_predictions)
root_mean_squared_error(y_test, test_predictions)
```
